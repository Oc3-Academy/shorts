## **Como criar *tidy data* no `pandas`?**

```python
import pandas as pd
import seaborn as sns
```

```python
pg_df = sns.load_dataset("penguins")
```

```python
pg_df
```

## **Como o `melt` functiona?**

```python
pd.melt(
    pg_df, 
    id_vars=["species", "island"], 
    value_vars=["bill_length_mm", "bill_depth_mm"],
)
```
