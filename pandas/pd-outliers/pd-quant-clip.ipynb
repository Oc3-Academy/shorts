## *Como remover outliers com `pandas`?*

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import fetch_openml
```

```python
housing = fetch_openml(name="house_prices", as_frame=True)
housing_df = pd.DataFrame(housing.data)
```

```python
housing_df
```

```python
body_mass = housing_df["LotFrontage"]
body_mass.clip(
    lower=body_mass.quantile(0.90),
    upper=body_mass.quantile(0.95)
)
```
