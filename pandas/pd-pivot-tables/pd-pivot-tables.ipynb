## **Como Utilizar o `pivot_table` do pandas?**

```python
import seaborn as sns
import pandas as pd
```

```python
pg_df = sns.load_dataset("penguins")
pg_df
```

## **Como functiona o `pivot_table` do pandas?**

![groupby](../../image/pivot_table.svg)

```python
pg_df.pivot_table(
    index=["species", "island"],
    values=["bill_length_mm"],
    aggfunc=[len, sum],
    margins=True
)
```
